```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars_df = pd.read_csv("mtcars.csv")
cars_df
```

```python
cars_df['mpg'].plot(kind = 'hist')
plt.xlabel("mpg")
```

```python
trans = cars_df['am'].value_counts()
print(trans)
trans.plot.bar()
plt.xlabel("Transmission")
```

```python
cars_df['mpg'].plot(kind = 'box')
```

```python
plt.scatter(cars_df['mpg'],cars_df['wt'],marker = '*')
plt.xlabel("mpg")
plt.show()
```
